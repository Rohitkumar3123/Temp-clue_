```python
from IPython import get_ipython
from IPython.display import display
# %% installing requirements
!pip install --upgrade --force-reinstall torch torchvision torchaudio --index-url https://download.pytorch.org/whl/cu118
!apt-get update && apt-get install -y tesseract-ocr

# %% importing liabraries
import cv2
import pytesseract
from transformers import pipeline, BartTokenizer, BartForConditionalGeneration


# pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'  # Update to your tesseract path if needed

# === STEP 1: OCR Image to Text ===
def extract_text_from_image(image_path):
    image = cv2.imread(image_path)
    if image is None:
        raise FileNotFoundError(f"Could not load image at {image_path}")

    grayscale = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    thresh = cv2.threshold(grayscale, 127, 255, cv2.THRESH_BINARY)[1]
    text = pytesseract.image_to_string(thresh)
    return text.strip()

# === STEP 2: Use LLM (BART) to Summarize Extracted Text ===
def summarize_text_with_llm(text):
    model_name = "facebook/bart-large-cnn"
    tokenizer = BartTokenizer.from_pretrained(model_name)
    model = BartForConditionalGeneration.from_pretrained(model_name)

    inputs = tokenizer.encode(text, return_tensors="pt", max_length=1024, truncation=True)
    summary_ids = model.generate(inputs, max_length=60, min_length=10, num_beams=4, early_stopping=True)
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary

# === STEP 3: Run Everything ===
if __name__ == "__main__":
    image_path = "/image2.webp"  # Changed path to /image2.webp
    print("üîç Extracting text from image...")
    extracted_text = extract_text_from_image(image_path)
    print("\nüìÉ Extracted Text:\n", extracted_text)

    if extracted_text:
        print("\nüß† Summarizing with LLM...")
        summary = summarize_text_with_llm(extracted_text)
        print("\nüìù Summary:\n", summary)
    else:
        print("‚ö†Ô∏è No text detected to summarize.")
```
